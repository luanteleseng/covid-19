## Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading Data

```python
df = pd.read_csv('./owid-covid-data.csv')
df.head()
```

```python
df_brazil = df.loc[df['location'] == 'Brazil']
df_brazil.head()
```

```python
df_brazil.shape
```

```python
df_brazil.info()
```

```python
df_brazil_vac = df_brazil.loc[df['new_vaccinations'] >= 1]
df_brazil_vac[['date','new_vaccinations']]
```

## Data Visualization

```python
# new_vaccinations
plt.figure(figsize=(20,6))
plt.title('New Vaccinations')

plt.plot(df_brazil_vac['date'], df_brazil_vac['total_vaccinations'])
plt.show()
```
